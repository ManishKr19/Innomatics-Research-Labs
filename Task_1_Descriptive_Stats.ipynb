#### **1.) Measure of Center**
  - Mean
  - Median
   

1.   **Mode**


**1.A) Mean: Mean is an average of all the numbers.**



The steps required to calculate a mean are:
  - sum up all the values of a target variable in the dataset
  - divide the sum by the number of values

i.e data = [10, 30, 25, 15, 40]
    
    Mean = (10 + 30 + 25 + 15 + 40)/5 = 24


```python

df.mean()
```

**1.B) Median: Median is the middle value of a sorted list of numbers.**

The steps required to get a median from a list of numbers are:
  - sort the numbers from smallest to highest
  - the value in the middle position is the median, if the list has an odd number of values
  - the average of the two values in the middle will be the median, if the list has an even number of values

i.e data = [30, 45, 55, 67, 87, 94, 102, 124]

    Median = (67+87)/2 = 77.0
    

```python
 
 df.median()
 
```

**1.1) Mode: Mode is the element having maxiumum frequency of occurance in a series.**

The steps required to get a median from a list of numbers are:
  - count the occurance of every element
  - the element with maximum occurance is mode. 

i.e data = [30, 45, 30, 87, 94, 102, 124]

    Mode = 30 (with max. count 2)
    

```python

df.mode() 
```

#### **2.) Measure of Variation**
  - Standard Deviation
  - Variartion

**2.A) Standard deviation (SD): The SD is just a measurement to tell how a set of values spread out from their mean.**

The steps to calculate SD are as follows:
  - calculate the mean of a dataset
  - for each number in the dataset, subtract it with the mean
  - square the difference obtained from Step 2
  - sum up the results from Step 3
  - divide the summation from Step 4 by the number of values in the dataset minus one
  - square root the result from Step 5

```python

df.std()
```

#### **4.) Normal Distribution:**
**It is a probability distribution that is symmetric about the mean, showing that data near the mean are more frequent in occurrence than data far from the mean.**


**KEY TAKEAWAYS**
  - A normal distribution is the proper term for a probability bell curve.
  - In a normal distribution the mean is zero and the standard deviation is 1.
  - It has zero skew and a kurtosis of 3.
  - Normal distributions are symmetrical, but vicevera is not true.

  Normal.jpg

```python
import matplotlib.pyplot as plt


mu, sigma = df['Annual_HH_Income'].mean(), df['Annual_HH_Income'].std()
s = np.random.normal(mu, sigma, 1000)

# Create the bins and histogram
count, bins, ignored = plt.hist(s,20)

# Plot the distribution curve
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) ), linewidth=3, color='y')
plt.show()
```

**4.A) Features of Normal Distribution:**
  - ***The mean, mode and median are all equal.***
  - The curve is symmetric at the center (i.e. around the mean, Œº).
  - Exactly half of the values are to the left of center and exactly half the values are to the right.
  - The total area under the curve is 1.

  standard-normal-distribution.jpg

**Positively skewed (Right-skewed) distribution:** 

Most values are clustered around the left tail of the distribution while the right tail of the distribution is longer.

Positively-Skewed-Distribution.webp

**Negatively skewed (Left-skewed) distribution:**

More values are concentrated on the right side (tail) of the distribution graph while the left tail of the distribution graph is longer.

#### **QQ Plot**

Graphical tool to help us assess if a set of data plausibly came from some theoretical distribution such as a Normal or exponential.

#### **Box Plot (whisker plot)**

A type of chart often used in explanatory data analysis. Box plots visually show the distribution of numerical data and skewness through displaying the data quartiles (or percentiles) and averages.

450px-Michelsonmorley-boxplot.png
