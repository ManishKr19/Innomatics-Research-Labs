```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sales_df = pd.read_csv(r'/content/BigMartSalesData.csv', parse_dates=['InvoiceDate'])
```

```python
sales_df.head()
```

```python
sales_df.info()
```

```python
sales_2011 = sales_df[sales_df['Year'] == 2011]['Amount'].values
date_2011 = sales_df[sales_df['Year'] == 2011]['InvoiceDate'].values
```

```python
plt.figure(figsize = (15,10))
plt.plot(date_2011, sales_2011)
```

```python
sales_2011 = sales_df[sales_df['Year'] == 2011]
salesByMonth_11 = sales_2011.groupby('Month')['Amount'].sum()
```

```python
months = salesByMonth_11.index
sales = salesByMonth_11.values
```

```python
plt.figure(figsize = (15,10))
plt.plot(months, sales, marker = 's', linewidth = 2)
```

***February 2011 has lowest sales***

```python
plt.figure(figsize = (15,10))
plt.bar(months, sales, width = 0.5)
```

```python
fig, ax = plt.subplots(figsize = (15,10))
ax.bar(months, sales,width = 0.5)

for i, txt in enumerate(sales):
    ax.annotate(txt, (months[i], sales[i]))
```

**yes, bar chart is simple and easy to understand.**
